# Intro to Agent Chat API with First Assistant Agent
Notebook focuses on creating an agent using the `Assistant Agent` API

```python
from autogen_agentchat.agents import AssistantAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
#print(f"API KEY: {api_key}")
```

```python
# Load the "brain"
model_client = OpenAIChatCompletionClient(model = 'gpt-4o', api_key=api_key)
```

# Create First Assistant AGent
The `Assistant Agent` is a a versatile agent for conversions, powered by AgentChat

```python
assistant = AssistantAgent(name = "assistant",model_client=model_client,description='A basic first agent')
```

# Test the Assistant Agent

```python
result = await assistant.run(task="what's the capital of France ?")
print(result)
```
